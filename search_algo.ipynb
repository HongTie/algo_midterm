A search problem involves a list of integers S and an integer x, and the task is to determine whether x is present in S or not. Our objective is to evaluate and compare the effectiveness of three search algorithms:  linear search, binary search, and Fibonacci search. We will start by generating a list S consisting of 10 randomly selected integers, along with another integer x. We will then execute all three algorithms on this list and record their respective execution times. By performing the task 5 times and calculating the total duration, we can obtain the mean execution time for each algorithm. To accomplish our objective, we will repeat the above process 100 times while increasing the size of the list S by 10 for each iteration. Therefore, we will execute the three search algorithms and record their average execution times for n = 10, 20, 30,.... up to 1000.

1. write code
2. Present a line chart depicting the average execution time for each algorithm with n values ranging from 100 to 10000.

```python
import random
import time

# Function for linear search
def linear_search(S, x):
    for i in range(len(S)):
        if S[i] == x:
            return True
    return False

# Function for binary search
def binary_search(S, x):
    low = 0
    high = len(S) - 1
    while low <= high:
        mid = (low + high) // 2
        if S[mid] == x:
            return True
        elif S[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return False

# Function for Fibonacci search
def fibonacci_search(S, x):
    fib1 = 0
    fib2 = 1
    fib3 = fib1 + fib2
    while fib3 < len(S):
        fib1 = fib2
        fib2 = fib3
        fib3 = fib1 + fib2
    offset = -1
    while fib3 > 1:
        i = min(offset+fib1, len(S)-1)
        if S[i] < x:
            fib3 = fib2
            fib2 = fib1
            fib1 = fib3 - fib2
            offset = i
        elif S[i] > x:
            fib3 = fib1
            fib2 = fib2 - fib1
            fib1 = fib3 - fib2
        else:
            return True
    #if fib2 and S[offset+1] == x:
    #    return True
    return False

for n in range(10, 1010, 10):
    # Generate a list of n random integers
    S = random.sample(range(n), n)
    x = random.randint(0, n-1)

    # Perform each search algorithm 5 times and calculate the total duration
    linear_time = 0
    binary_time = 0
    fibonacci_time = 0 
    num_trials = 5
    for i in range(num_trials):
        start_time = time.time()
        linear_search(S, x)
        end_time = time.time()
        linear_time += end_time - start_time
    
        S.sort()
        start_time = time.time()
        binary_search(S, x)
        end_time = time.time()
        binary_time += end_time - start_time
    
        start_time = time.time()
        fibonacci_search(S, x)
        end_time = time.time()
        fibonacci_time += end_time - start_time

    # Calculate the mean execution time for each algorithm
    linear_mean_time = linear_time / num_trials
    binary_mean_time = binary_time / num_trials
    fibonacci_mean_time = fibonacci_time / num_trials

    # Print the mean execution time for each algorithm
    print("when n = ", n)
    print("Linear search mean execution time:", linear_mean_time)
    print("Binary search mean execution time:", binary_mean_time)
    print("Fibonacci search mean execution time:", fibonacci_mean_time)



```
